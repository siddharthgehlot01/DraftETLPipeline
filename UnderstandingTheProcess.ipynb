```python
data_subset = data_subset.groupby(['ISIN','Date']).agg(avg_min_price = ('MinPrice','mean'),
                                                       avg_max_price = ('MaxPrice','mean'),
                                                        avg_trade_volume=('TradedVolume','mean'),
                                                      total_trade_number=('NumberOfTrades','sum') )





```
